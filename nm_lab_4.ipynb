```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def newtons_raphson(y,y_p,x0,max_iter=40,tol=1e-6):
  x_lists=[]
  for _ in range(max_iter):
    x1=x0-(y(x0)/y_p(x0))
    if(abs(x0-x1))<tol:
      break
    x_lists.append(x0)
    x0=x1;
  plt.scatter(x_lists,y(x_lists),s=20,label="Newtons Raphson Approximation",color='red')
  plt.axhline(color='green',label="y=0 line")
  plt.plot(np.arange(-12,12,0.01),y(np.arange(-12,12,0.01)),label="Orignial Function")
  plt.xlabel("x")
  plt.grid()
  plt.title("Newtons Raphson Approximation 080BCT060")
  plt.ylabel("y")
  plt.legend(loc='best')
  print(f"The value of x is {x_lists[-1]}")
  plt.show()
  return x_lists
```

```python
def y2(x):
  return np.exp(x)

def y2_p(x):
  return np.exp(x)


print("\nTest Case 2:")
sol=newtons_raphson(y2, y2_p, 3.4)
```

```python
def least_sqr_linear(x_ar,y_ar):
  A=np.array([[x_ar.size,np.sum(x_ar)],
              [np.sum(x_ar),np.sum(x_ar**2)]])
  B=np.array([np.sum(y_ar),np.sum(x_ar*y_ar)])
  coeff=np.linalg.solve(A,B)
  plt.scatter(x_ar,y_ar,label="Original Data",color='red',s=22)
  plt.plot(x_ar,coeff[1]*x_ar+coeff[0],color='green',linewidth=2,label="Least Square Approximation")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.grid()
  plt.title("Least Square Approximation 080BCT060")
  plt.legend(loc='best')

  return (coeff)
```

```python
f1=lambda x:3*x+13+np.random.normal(0,3,len(x))
x_test=np.arange(-1,10,1)
y_test=f1(x_test)

least_sqr_linear(x_test,y_test)
```

```python

print("\nTest Case 2:")
x_test2 = np.arange(-1, 10, 0.1)
y_test2 = np.array(x_test2)+32+np.random.normal(0,1,len(x_test2))
coeff2 = least_sqr_linear(x_test2, y_test2)
print(f"Least square coefficients: {coeff2}")
```

```python

print("\nTest Case 3:")
x_test3 = np.arange(-5, 5, 0.1)
y_test3 = 2 + 2*x_test3 + 5+np.random.normal(0,4,len(x_test3))
coeff3 = least_sqr_linear(x_test3, y_test3)
print(f"Least square coefficients: {coeff3}")
```

```python

def exponential_curve_fit(x_ar,y_ar):
  Y=np.log(y_ar)
  A=np.array([[x_ar.size,np.sum(x_ar)],
              [np.sum(x_ar),np.sum(x_ar**2)]])
  B=np.array([np.sum(Y),np.sum(x_ar*Y)])
  coeff=np.linalg.solve(A,B)
  plt.scatter(x_ar,y_ar,label="Original Data",color='red',s=22)
  a=np.exp(coeff[0])
  b=(coeff[1])

  plt.plot(x_ar,a*np.exp(b*x_ar),color='green',linewidth=2,label="Exponential Approximation")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.grid()
  plt.title("Exponential curve fit 080BCT060")
  plt.legend(loc='best')

  return (a,b)
```

```python
x_ar=np.arange(10,20,.5)
y_ar=np.exp(x_ar*0.1)*21+np.random.normal(0,4,len(x_ar))
exponential_curve_fit(x_ar,y_ar)
```

```python
def poly_curve_fit(X_ar,y_ar,degree=3):
  A=np.zeros((degree+1,degree+1))
  B=np.zeros((degree+1))

  for i in range(degree+1):
    for j in range(degree+1):
      A[i][j]=np.sum(X_ar**(i+j))
    B[i]=np.sum(y_ar*(X_ar**i))

  coeffs=np.linalg.solve(A,B)
  y=np.polyfit(X_ar,y_ar,degree)
  plt.scatter(X_ar,y_ar,label="Original Data",color='red',s=22)
  plt.plot(X_ar,np.polyval(y,X_ar),color='green',linewidth=2,label="Polynomial Approximation")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.grid()
  plt.title("Polynomial curve fit 080BCT060")
  plt.legend(loc='best')
  plt.show()
```
